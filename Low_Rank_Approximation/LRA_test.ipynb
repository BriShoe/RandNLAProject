```python
from approximate_multiplication import LRA, LRA_r
import numpy as np
import pandas as pd
import time

svd_times = []
approx_times = []
for i in range(4, 14):
    A = np.random.normal(0, 1, size=(2**i, 2**i))
    eps, k = 1e-1, 10
    start1 = time.time()
    estimate = LRA(A, k, eps, s=100)
    print(i)
    print("Frobenius Distance:", np.linalg.norm(A - estimate))
    end1 = time.time()
    print("Time:", end1 - start1)
    approx_times.append(end1 - start1)
    start2 = time.time()
    best_estimate = LRA_r(A, k)
    print("Frobenius Distance:", np.linalg.norm(A - best_estimate))
    end2 = time.time()
    print("Time:", end2 - start2)
    svd_times.append(end2-start2)
```

```python
import matplotlib.pyplot as plt

plt.plot(np.arange(4, 14), svd_times, label="LRA via SVD")
plt.plot(np.arange(4, 14), approx_times, label="Approximate LRA")
plt.legend()
plt.xlabel("$log_2$ of number of rows and columns of $A$")
plt.ylabel("Time (seconds)")
plt.title("Time elapsed for Low Rank Approximation")
plt.show()
```
