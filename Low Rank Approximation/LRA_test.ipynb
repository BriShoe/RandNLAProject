```python
from approximate_multiplication import LRA, LRA_r
import numpy as np
import pandas as pd
import time

A = np.random.normal(0, 1, size=(10000, 5000))
eps, k = 1e-1, 100 
start1 = time.time()
estimate = LRA(A, k, eps, s=100)
print(np.linalg.norm(A - estimate))
end1 = time.time()
print(end1 - start1)
start2 = time.time()
best_estimate = LRA_r(A, k)
print(np.linalg.norm(A - LRA_r(A, k)))
end2 = time.time()
print(end2 - start2)
```
