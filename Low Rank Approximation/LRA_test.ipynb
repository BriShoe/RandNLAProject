```python
from approximate_multiplication import LRA, LRA_r
import numpy as np
import pandas as pd

A = np.random.normal(0, 1, size=(10000, 5000))
eps, k = 1e-1, 10 
estimate = LRA(eps, A, k, s=100)
best_estimate = LRA_r(A, k)
print(np.linalg.norm(A - estimate))
print(np.linalg.norm(A - LRA_r(A, k)))
```
